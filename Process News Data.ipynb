```python
def make_lat(point):
    return point[0]
def make_lon(point):
    return point[1]
```

```python
def make_points(row):
    from shapely.geometry import Point
    row['Point'] = Point(row['lon'],row['lat'])
    return row
```

```python
def return_province_id(row):
    point = row['Point']
    for province in boundaries.iterrows():
        if point.within(province[1]['geometry']):
            row['province'] = province[1]['ID_1']
    return row
        
```

```python
boundaries = gpd.read_file('Shapefiles/IDN_adm/IDN_adm1.shp')

boundaries
```

```python
locations = pd.read_pickle('locations.pickle')
```

```python
locations = locations.dropna()

locations['lat'] = locations['latlon'].apply(make_lat)
locations['lon'] = locations['latlon'].apply(make_lon)

locations = locations.apply(make_points, axis=1)

locations = locations.apply(return_province_id, axis=1)
```

```python
locations
```

```python
news = pd.read_excel('imm-news.xlsx')
news['news_date'] = pd.to_datetime(news['news_date']).dt.date
```

```python
def get_province_for_outlet(outlet_name):
    try:
        return locations[locations['outlet_name'] == outlet_name]['province'].values[0]
    except:
        return None
```

```python
news['province'] = news['media_displayName'].apply(get_province_for_outlet)
```
